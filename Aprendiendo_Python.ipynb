Acumulador de suma

````python
print("suma iterativa")
numero= 1
acumulador_suma=0
i=0
while i <100  :
  acumulador_suma+= numero
  numero= numero +1
  i=i+1

print(f"{acumulador_suma} \n")
````

````python
print("***menu interactivo para todos ***")
salir = False
while not salir:
  print("""
  1. crear cuenta
  2. actualizar cuenta
  3. salir
  """)
  opcion = int(input(f"escoge una opcion"))
  if opcion == 1:
    print("Creando tu cuenta...")

  elif opcion ==2:
    print("actualizando tu cuenta...")

  elif opcion ==3:
    print("Saliendo...")
    salir=True
  else:
    print("escoge una opcion valida")
````

 # Listas

````python
lista_prueba=[7,"hola mundo", 23.32, "tu"]
print(lista_prueba[2]) #mostrar

print("====================AÑADIR=====================")
#añade a la lista
lista_prueba.append("papi tu sabes")
print(lista_prueba)

#con esta se pueden añadir mas de 1 elemento
lista_prueba.extend(["es que a falta de 1", "dos"])
print(lista_prueba)

#añade en una posicion en especifico
lista_prueba.insert(0,"ahora yo soy el primero")
print(lista_prueba)
````

````python
print("====================ELIMINAR===================")

#elimina las coinsidencias
lista_prueba.remove("tu")

#eliminad por indice en la lista
lista_prueba.pop(1)
print(lista_prueba)

#eliminar el ultimo elemento
lista_prueba.pop()
print(lista_prueba)

#elimiar la lista completa
lista_prueba.clear()
print(lista_prueba)

````

````python
print("====================EDITAR===================")

#editar un elemento de la lista
lista_prueba[3] = "no se"
print(lista_prueba)

#elimiar la lista completa
lista_prueba[3:5] = [12,45,5]
print(lista_prueba)

for puebe in lista_prueba:
  if puebe == 12:
    print ("aquita")
    print(puebe)
    print(" ")
  else:
    print(puebe)


````

Tuplas

````python
#la tupla se crea y no se puede editar, elimirar o añadir elementos

#pero si se pueden comvertir de una a otra
print("De tuplas a listas")
frutas = ("manzana", "pera", "uva")

lista = list(frutas)

print(lista)

print("De listas a tuplas")
lista = ["manzana", "pera", "uva"]

tupla = tuple(lista)

print(tupla)
````
