````python
class Persona:

  def inicializar_persona(self,nombre,apellido):
    #creamos atributos de la clase
    self.nombre = nombre
    self.apellido = apellido

  def mostrar_persona(self):
    print(f""" Persona:
    nombre {self.nombre}
    apellido {self.apellido}
    """)

#tenemos que inicializar osea crear a la persona y
#añadirle atributos
Persona1 = Persona()
Persona1.inicializar_persona("Esteban","Espitia")
print(Persona1.mostrar_persona())
````
